```python
import pandas as pd
import matplotlib.pyplot as mapo
import seaborn as sebo

df=pd.read_csv('mtcars.csv')
```

```python
best_mpg_car = df.loc[df['mpg'].idxmax()]
print("Car with the best mpg:\n", best_mpg_car)
```

```python
worst_mpg_car = df.loc[df['mpg'].idxmin()]
print("Car with the worst mpg:\n", worst_mpg_car)
```

```python
best_hp_car = df.loc[df['hp'].idxmax()]
print("Car with the best horsepower:\n", best_hp_car)
```

```python
#Find 5 number summary of displacement
displacement_summary = df['disp'].describe()
print("5 number summary of displacement:\n", displacement_summary)
```

```python
median_hp = df['hp'].median()
print("Median horsepower:", median_hp)
```

```python
average_mpg_by_am = df.groupby('am')['mpg'].mean()
print("Average mpg for manual vs. automatic cars:\n", average_mpg_by_am)
```

```python
#Draw a histogram of miles per gallon
mapo.figure(figsize=(10, 6))
sebo.histplot(df['mpg'], bins=10, kde=True)
mapo.title('Histogram of Miles Per Gallon')
mapo.xlabel('Miles Per Gallon')
mapo.ylabel('Frequency')
mapo.show()
```

```python
#Boxplot of mpg for each cylinder type
mapo.figure(figsize=(10, 6))
sebo.boxplot(x='cyl', y='mpg', data=df)
mapo.title('Boxplot of MPG for Each Cylinder Type')
mapo.xlabel('Number of Cylinders')
mapo.ylabel('Miles Per Gallon')
mapo.show()
```

```python
#Create a crosstab displaying count of automatic vs. manual cars
crosstab_am = pd.crosstab(df['am'], columns='count')
print("Crosstab of automatic vs. manual cars:\n\n", crosstab_am)
```

```python
#create a crosstab displaying count of “am vs cyl”
crosstab_am_cyl = pd.crosstab(df['am'], df['cyl'])
print("Crosstab of am vs cyl:\n\n", crosstab_am_cyl)
```

```python
#correlation between the weight of the car and mpg
correlation_wt_mpg = df['wt'].corr(df['mpg'])
print("Correlation between weight and mpg:", correlation_wt_mpg)
```
